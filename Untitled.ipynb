```python
corr = df_heat.corr()
```

```python
corr
```

```python

corr.applymap(lambda x: f"{x:.1f}")
```

```python
corr.shape
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
mask
```

```python

fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(corr, mask=mask, ax=ax,cmap='inferno', annot=True,fmt=".1f",linewidth=1,cbar=True, cbar_kws={'location': 'right', 'ticks': [-0.08, 0.00 ,0.08 ,0.16, 0.24],'aspect':30});
ax.set(xlabel="", ylabel="")
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.grid(False)

fig.savefig('part.jpeg')
```

```python
np.arange(len(corr.columns))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

# Create heatmap with Seaborn and store the heatmap object
heatmap = sns.heatmap(corr, mask=mask, ax=ax, cmap="YlGnBu", annot=True, fmt=".1f", linewidth=1)

# Access the color bar object associated with the heatmap
cbar = heatmap.collections[0].colorbar

# Set custom color bar range and ticks
vmin = -0.08
vmax = 0.32
cbar.set_ticks(np.arange(vmin, vmax , 0.08))  # Adjust ticks with step 0.08
cbar.set_ticklabels([f'{x:.2f}' for x in np.arange(vmin, vmax , 0.08)])  # Format tick labels

# Adjust the height and width of the color bar
cbar.ax.figure.set_size_inches(15, 10)  # Set the size of the color bar (width, height)
```
